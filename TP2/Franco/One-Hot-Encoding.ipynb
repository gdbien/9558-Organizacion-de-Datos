```python
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
#Cargo los sets
X_train = pd.read_csv('data/X_train.csv', dtype={'id':'int',\
                                           'tipodepropiedad':'category',\
                                           'ciudad':'category', 'provincia':'category',\
                                           'antiguedad':'float', 'habitaciones':'float',\
                                           'garages':'float','banos':'float',\
                                           'metroscubiertos':'float', 'metrostotales':'float',\
                                           'idzona':'float',\
                                           'gimnasio':'bool', 'usosmultiples':'bool',\
                                           'piscina':'bool', 'escuelascercanas':'bool',\
                                           'centroscomercialescercanos':'bool'\
                                           }, index_col='id')
X_test = pd.read_csv('data/X_test.csv', dtype={'id':'int',\
                                           'tipodepropiedad':'category',\
                                           'ciudad':'category', 'provincia':'category',\
                                           'antiguedad':'float', 'habitaciones':'float',\
                                           'garages':'float','banos':'float',\
                                           'metroscubiertos':'float', 'metrostotales':'float',\
                                           'idzona':'float',\
                                           'gimnasio':'bool', 'usosmultiples':'bool',\
                                           'piscina':'bool', 'escuelascercanas':'bool',\
                                           'centroscomercialescercanos':'bool'\
                                           }, index_col='id')
```

```python
X_submit = pd.read_csv('data/X_submit.csv', dtype={'id':'int',\
                                           'tipodepropiedad':'category',\
                                           'ciudad':'category', 'provincia':'category',\
                                           'antiguedad':'float', 'habitaciones':'float',\
                                           'garages':'float','banos':'float',\
                                           'metroscubiertos':'float', 'metrostotales':'float',\
                                           'idzona':'float',\
                                           'gimnasio':'bool', 'usosmultiples':'bool',\
                                           'piscina':'bool', 'escuelascercanas':'bool',\
                                           'centroscomercialescercanos':'bool'\
                                           }, index_col='id')
```

```python
cat_columns = ['tipodepropiedad','provincia','ciudad']
```

```python
df_train = pd.get_dummies(X_train, prefix_sep="__", columns=cat_columns)
```

```python
cat_dummies = [col for col in df_train 
               if "__" in col 
               and col.split("__")[0] in cat_columns]
```

```python
processed_columns = list(df_train.columns[:])
```

```python
df_test = pd.get_dummies(X_test, prefix_sep="__", 
                                   columns=cat_columns)
```

```python
# Remove additional columns
for col in df_test.columns:
    if ("__" in col) and (col.split("__")[0] in cat_columns) and col not in cat_dummies:
        print("Removing additional feature {}".format(col))
        df_test.drop(col, axis=1, inplace=True)
```

```python
for col in cat_dummies:
    if col not in df_test.columns:
        print("Adding missing feature {}".format(col))
        df_test[col] = 0
```

```python
df_test = df_test[processed_columns]
```
