```python
#Cargo los sets
X_train_cols = pd.read_csv('data/X_train.csv', dtype={'id':'int',\
                                           'tipodepropiedad':'category',\
                                           'ciudad':'category', 'provincia':'category',\
                                           'antiguedad':'float', 'habitaciones':'float',\
                                           'garages':'float','banos':'float',\
                                           'metroscubiertos':'float', 'metrostotales':'float',\
                                           'idzona':'float',\
                                           'gimnasio':'bool', 'usosmultiples':'bool',\
                                           'piscina':'bool', 'escuelascercanas':'bool',\
                                           'centroscomercialescercanos':'bool'\
                                           }, index_col='id')
X_test_cols = pd.read_csv('data/X_test.csv', dtype={'id':'int',\
                                           'tipodepropiedad':'category',\
                                           'ciudad':'category', 'provincia':'category',\
                                           'antiguedad':'float', 'habitaciones':'float',\
                                           'garages':'float','banos':'float',\
                                           'metroscubiertos':'float', 'metrostotales':'float',\
                                           'idzona':'float',\
                                           'gimnasio':'bool', 'usosmultiples':'bool',\
                                           'piscina':'bool', 'escuelascercanas':'bool',\
                                           'centroscomercialescercanos':'bool'\
                                           }, index_col='id')
```

```python
y_train_log = np.log1p(y_train)
y_test_log = np.log1p(y_test)
```

```python
X_train_cols = X_train_cols.iloc[:,1].to_frame()
X_train_cols['precio_log'] = y_train_log
```

```python
X_test_cols = X_test_cols.iloc[:,1].to_frame()
X_test_cols['precio_log'] = y_test_log
```

```python
X_submit_cols = X_submit_cols.iloc[:,1].to_frame()
X_submit_cols['precio_log'] = y_train_log
```

```python
X_submit_cols
```

```python
ce_bin = ce.BinaryEncoder(cols = ['ciudad'])
X_train_cols = ce_bin.fit_transform(X_train_cols,X_train_cols['precio_log'])
X_train_cols.drop(columns=['precio_log'],inplace=True)
```

```python
X_test_cols = ce_bin.transform(X_test_cols,X_test_cols['precio_log'])
X_test_cols.drop(columns=['precio_log'],inplace=True)
```

```python
X_submit_cols = ce_bin.transform(X_submit_cols,X_submit_cols['precio_log'])
X_submit_cols.drop(columns=['precio_log'],inplace=True)
```

```python
X_train.drop(X_train.filter(like='ciudad__',axis=1).columns,axis=1,inplace=True)

X_test.drop(X_test.filter(like='ciudad__',axis=1).columns,axis=1,inplace=True)
```

```python
X_submit.drop(X_submit.filter(like='ciudad__',axis=1).columns,axis=1,inplace=True)
```
